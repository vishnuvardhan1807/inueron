```python
def clean_FlightNumber(df):
    for i in range(1, len(df["FlightNumber"])):
        df["FlightNumber"][i] = df["FlightNumber"][i - 1] + 10
clean_FlightNumber(df)
df["FlightNumber"] = df["FlightNumber"].astype(int)
df.head()
```

<pre>
5. In the RecentDelays column, the values have been entered into the
DataFrame as a list. We would like each first value in its own column, each

second value in its own column, and so on. If there isn't an Nth value, the value
should be NaN.
Expand the Series of lists into a DataFrame named delays, rename the columns
delay_1, delay_2, etc. and replace the unwanted RecentDelays column in df
with delays.
</pre>

```python
delays = list(df2["RecentDelays"].values)
lengths = []
for list_ in delays:
    lengths.append(len(list_))
```

```python
max_length = max(lengths)
max_length
```

The max length is 3 so the maximum number additional columns we need is 3

```python
delay_1 = []
delay_2 = []
delay_3 = []

for i in delays:
    if len(i) == 3:
        delay_1.append(i[0])
        delay_2.append(i[1])
        delay_3.append(i[2])
    
    elif len(i) == 2:
        delay_1.append(i[0])
        delay_2.append(i[1])
        delay_3.append("NaN")
    
    elif len(i) == 1:
        delay_1.append(i[0])
        delay_2.append("NaN")
        delay_3.append("NaN")
    else:
        delay_1.append("NaN")
        delay_2.append("NaN")
        delay_3.append("NaN")
        
```

```python
df2["Delay_1"] = delay_1
df2["Delay_2"] = delay_2
df2["Delay_3"] = delay_3
df2.head()
```
