```python
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D
```

```python
data = load_iris()
df = pd.DataFrame(data['data'], columns=data['feature_names'])
df['targets'] = data['target']
df.head()
```

```python
# Divide into dependent and independent features
x = df.iloc[:, :].values
y = df.iloc[ :, 4].values
```

```python
# check the shape of variables
print(x.shape)
print(y.shape)
```

```python
# standardizing the features
scaler = StandardScaler()
x_std = scaler.fit_transform(x)
```

```python
# Transforming into three dimensional
pca = PCA(n_components=3)
X = pca.fit_transform(x_std)
```

```python
# Visualizng
plt.style.use('fivethirtyeight')
figure = plt.figure(1, figsize=(5, 10))
#plt.clf()
ax = Axes3D(figure, rect=[0, 0, 95, 1])
ax.scatter(X[:,0], X[:,1], X[:,2], c=y)
ax.set_title("3D visualization")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
plt.show()
```
